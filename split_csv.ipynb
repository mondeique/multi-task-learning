# import library

```python
import pandas as pd
```

# Category csv handling

```python
df = pd.read_csv('./bag_image_csv/choco_categories_categories_v3.csv')
```

```python
df
```

```python
## 중복 제거

df = df.drop_duplicates(['cropped_image_id'])
```

```python
df
```

```python
## index 내림차순으로 정리

df = df.reset_index(drop=True)
```

```python
df
```

```python
df.insert(2, "filename", "NONE") 
```

```python
df
```

```python
is_categories_list = []
for i in range(len(df)):
    is_categories_list.append(df['cropped_image_id'][i])
```

```python
len(is_categories_list)
```

# Cropped image handling

```python
df2 = pd.read_csv('./bag_image_csv/choco_categories_croppedimage_v3.csv')
```

```python
## category 있는 것들만 남기기

for i in range(len(df2)):
    if df2['id'][i] not in is_categories_list:
        df2 = df2.drop([i])
```

```python
df2
```

```python
df2 = df2.reset_index(drop=True)
```

```python
df2
```

```python
## Test image 있는지 확인하기

for i in range(len(df2)):
    if df2['image'][i] == 'cropped-images-dev/0D6OZW9C9MC_1.jpg':
        print("YES")
```

## cropped image와 category id로 matching 시키기 

```python
for i in range(len(df2)):
    for j in range(len(df)):
        if df2['id'][i] == df['cropped_image_id'][j]:
            df['filename'][j] = df2['image'][i]
```

```python
df
```

## post-processing for making training csv

```python
for i in range(len(df)):
    df['filename'][i] = './data/bag_image/' + df['filename'][i].split('/')[-1] 
```

```python
df
```

```python
df = df.drop(columns = ['id', 'version', 'material', 'cropped_image_id'])
```

```python
df
```

```python
df = df[['filename','color','shape','opening_type', 'strap_type', 'pattern', 'carring_type', 'charm', 'decoration']]
```

```python
df
```

```python
df = df.replace({'pattern' : { 99 : 7}})
df = df.replace({'decoration' : { 99 : 10}})
df = df.replace({'charm' : { 99 : 5}})
df = df.replace({'strap_type' : { 99 : 6}})
```

```python
df
```

```python
## export to csv

df.to_csv('./training_bag_v3_csv', index=False)
```
